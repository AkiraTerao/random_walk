```
from sympy import *
init_printing(False)
var('n  k')
import numpy as np
import matplotlib.pyplot as plt
```

1D Random walk return to origin at n steps

```
step=100
prop_theo_d=[]
st_d=[]
for i in np.arange(1,int(step/2)+1,1):
 p=Sum(1/(2**(2*n))*factorial(2*n)/(factorial(n)*factorial(n)),(n,1,i)).evalf()
 prop_theo_d.append(p)
 st_d.append(int(2*i))

plt.plot(st_d,prop_theo_d)
plt.title('Theoretical calculation')
plt.ylabel('Cumulative probability')
plt.xlabel('Steps')
```

```
# V=1
num=1000000
step=100
prop_sim_1d=[]
st_s=[]
nums=[]
a=np.random.random((num,step))
a[a>=0.5]=1
a[a<0.5]=-1

for i in np.arange(2,step+2,2):
 sa=np.sum(a[:,:i],axis=1)
 ind=np.where(sa==0)
 c=np.size(ind)
 p=c/num
 prop_sim_1d.append(p)
 st_s.append(i)
cm_prop_sim_1d=np.cumsum(prop_sim_1d)
plt.plot(st_s,cm_prop_sim_1d)
plt.title('Numerical Simulation')
plt.ylabel('Cumulative probability')
plt.xlabel('Steps')
```

```
plt.plot(st_d,prop_theo_d)
plt.plot(st_s,cm_prop_sim_1d,'r--')
plt.title('Theoretical vs Simulation')
plt.ylabel('Cumulative probability')
plt.xlabel('Steps')
plt.legend(['Theory','Simulation'])
```

```
step=30
st_t=[]
prop2d_theo=[]
for i in np.arange(1,int(step/2)+1,1):
 p2=1/(4**(2*i))*Sum(factorial(2*i)/((factorial(k))**2*(factorial(i-k))**2),(k,0,i)).evalf()
 prop2d_theo.append(p2)
 st_t.append(int(i*2))
cm_prop2d_theo=np.cumsum(prop2d_theo)
plt.plot(st_t,cm_prop2d_theo)
plt.title('Theoretical Calculation')
plt.ylabel('Cumulative probability')
plt.xlabel('Steps')
```

```
num=10000
step=100
prop2d_sim=[]
st=[]
a=np.random.random((num,step))
a[a>0.5]=1
a[a<=0.5]=-1

b=np.random.random((num,step))
b[a>0.5]=1
b[a<=0.5]=-1

v=np.random.random((num,step))
v[v>0.5]=1
v[v<=0.5]=0
vnot=np.logical_not(v).astype(int)

x=a*v
y=b*vnot
for i in np.arange(2,step+2,2):
 sx=np.sum(x[:,:i],axis=1)
 sy=np.sum(y[:,:i],axis=1)

 ind=np.where((sx==0) & (sy==0))
 c=np.size(ind)
 p=c/num
 prop2d_sim.append(p)
 st.append(int(i))
cm_prop2d_sim=np.cumsum(prop2d_sim)
plt.plot(st,cm_prop2d_sim)
plt.title('Numerical Simulation')
plt.ylabel('Cumulative probability')
plt.xlabel('Steps')
```

```
plt.plot(st_t, cm_prop2d_theo)
plt.plot(st,cm_prop2d_sim,'r--')
plt.title('Theoretical vs Simulation')
plt.ylabel('Cumulative probability')
plt.xlabel('Steps')
plt.legend(['Theory','Simulation'])
plt.ylim([0,1.4])
```

3D random walk

```
step=34
st_t=[]
prop3d_theo=[]
for i in np.arange(1,int(step/2)+1,1):
 p3=1/(2**(2*i))*factorial(2*i)/(factorial(i))**2*Sum((1/3**(i)*factorial(i)/(factorial(k)*factorial(i-k)))**2*factorial(2*k)/(factorial(k))**2,(k,0,i)).evalf()
 prop3d_theo.append(p3)
 st_t.append(int(i*2))
cm_prop3d_theo=np.cumsum(prop3d_theo)
plt.plot(st_t,cm_prop3d_theo)
plt.title('Theoretical Calculation')
plt.ylabel('Cumulative probability')
plt.xlabel('Steps')
```

```
num=10000
step=100
prop3d_sim=[]
st_s=[]
a=np.random.random((num,step))
a[a>0.5]=1
a[a<=0.5]=-1

b=np.random.random((num,step))
b[a>0.5]=1
b[a<=0.5]=-1

b2=np.random.random((num,step))
b2[a>0.5]=1
b2[a<=0.5]=-1

vx=np.random.random((num,step))
vx[vx>2/3]=1
vx[vx<=2/3]=0

vxnot=np.logical_not(vx).astype(int)

v=np.random.random((num,step))
v[v>1/2]=1
v[v<=1/2]=0

vy=v*vxnot
vz=np.logical_not(v)*vxnot

x=vx*a
y=vy*b
z=vz*b2

for i in np.arange(2,step+2,2):
 sx=np.sum(x[:,:i],axis=1)
 sy=np.sum(y[:,:i],axis=1)
 sz=np.sum(z[:,:i],axis=1)

 ind=np.where((sx==0) & (sy==0) & (sz==0))
 c=np.size(ind)
 p3=c/num
 prop3d_sim.append(p3)
 st_s.append(int(i))
cm_prop3d_sim=np.cumsum(prop3d_sim)
plt.plot(st_s,cm_prop3d_sim)
plt.title('Numerical Simulation')
plt.ylabel('Cumulative probability')
plt.xlabel('Steps')
```

```
plt.plot(st_t,cm_prop3d_theo)
plt.plot(st_s,cm_prop3d_sim,'r--')
plt.title('Theoretical vs Simulation')
plt.ylabel('Cumulative probability')
plt.xlabel('Steps')
plt.legend(['Theory','Simulation'])
plt.ylim([0,1])
```
